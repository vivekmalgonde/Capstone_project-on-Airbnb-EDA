üí†**Minimum price is zero which is not possible .**

```python
df[df['price']==0]
```

```python
# A function for price_correction
def price_correction(x):
    if x==0:
        return 100
    else:
        return x
```

```python
df['price']=df['price'].apply(price_correction)
```

```python
df['price'].apply(price_correction)
```

```python
df[df['price']==0]
```

```python
df.describe()
```

üí†**In this data you can see that max minimum nights is 1250 it is also not 
relevant**

```python
def minimum_nights_correction(x):
  if x>365:
    return 365
  else:
    return x
```

```python
df['minimum_nights']=df['minimum_nights'].apply(minimum_nights_correction)
```

```python
df[df['minimum_nights']==365]
```

```python
df.describe()
```

##‚ö°**ANALYSING THE RELATION BETWEEN NEIGHBOURHOOD GROUP AND PRICE.**

####üí†We can observe that **Manhattan** is having the highest median price,followed by **Brooklyn**.

# ‚ñ∂Considering top 20 hosts

## By using above data we can plot a graph for easy visual understanding

```python

fig_5=sns.barplot(x='Host_id',y='counts',data=df_top)
fig_5.set_title('Hosts with the most listings in NYC')
fig_5.set_ylabel('Count of listings')
fig_5.set_xlabel('Host IDs')
fig_5.set_xticklabels(fig_5.get_xticklabels(), rotation=45)
plt.show()
```

From the above Bar Chart we come to know that 219517861 host_id have highest stays

# **‚ö° TOP 20 EXPENSIVE NEIGHBOURHOOD BASED ON THE AVERAGE PRICE**

```python
Expen_neigh = df.groupby(["neighbourhood"])['price'].aggregate(np.median).reset_index().sort_values(['price'],ascending=False)
Expen_neigh
```

```python

fig_6=sns.barplot(x='neighbourhood',y='price',data=Expen_neigh[:20])
fig_6.set_title('TOP 10 EXPENSIVE NEIGHBOURHOOD BASED ON THE AVERAGE PRICE')
fig_6.set_ylabel('Average Price')
fig_6.set_xlabel('NEIGHBOURHOOD')
fig_6.set_xticklabels(fig_6.get_xticklabels(), rotation=45)
```

#**‚ö° TOP 20 INEXPENSIVE NEIGHBOURHOOD BASED ON THE AVERAGE PRICE**

```python
fig_6=sns.barplot(x='neighbourhood',y='price',data=Expen_neigh[-20:])
fig_6.set_title('TOP 10 EXPENSIVE NEIGHBOURHOOD BASED ON THE AVERAGE PRICE')
fig_6.set_ylabel('Average Price')
fig_6.set_xlabel('NEIGHBOURHOOD')
fig_6.set_xticklabels(fig_6.get_xticklabels(), rotation=45)
```
